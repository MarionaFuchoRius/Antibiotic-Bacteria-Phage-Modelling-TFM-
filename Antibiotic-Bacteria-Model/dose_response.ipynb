```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["text.usetex"] = True
# Simulation of a  a Michaelis-Menten type response curve
A_vals = np.linspace(0, 50, 100)
E_max = 1.0
EC50 = 1.0  


response_curve = E_max * A_vals / (EC50 + A_vals)


plt.figure(figsize=(7, 5))
plt.plot(A_vals, response_curve, color='black', lw=2)
plt.xlabel(r'Antibiotic Concentration (A)', fontsize=14)
plt.ylabel(r'Effectiveness', fontsize=14)
#plt.title('Dose-Response Curve', fontsize=16)
plt.grid(False)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.savefig('effectiveness.pdf')
plt.show()
```
